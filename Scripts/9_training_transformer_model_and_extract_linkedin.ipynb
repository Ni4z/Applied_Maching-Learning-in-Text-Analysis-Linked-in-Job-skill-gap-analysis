# Prediction Job Data


```python
import pandas as pd
import spacy
```

```python
df_full = pd.read_csv('/content/drive/MyDrive/aml4ta/linkedIn_all_translated_final.csv')
df_full.head()
```

```python
df_data_scien = df_full[df_full["title"].str.contains("data scien")]
df_data_scien
```

```python
trained_nlp = spacy.load("/content/drive/MyDrive/aml4ta/transformer_model/model-best")
collected_skills_job = []
i= 0
for index, row in df_data_scien.iterrows():
    print(i,"/",len(df_data_scien))
    doc = trained_nlp(row["content"])
    for ent in doc.ents:
        collected_skills_job.append({
            "id": row["id"],
            "skill": ent.text.lstrip('-').rstrip('-'),
            "label": ent.label_,
        })
    i+=1
```

```python
df_skills_job = pd.DataFrame(collected_skills_job)
df_skills_job
```

```python
df_skills_job.to_json("/content/drive/MyDrive/aml4ta/transformer/linkedin_skills_data_scien.json", orient="records")
```

```python
df_hard_job = df_skills_job[df_skills_job["label"]=="HARD"] 
df_soft_job = df_skills_job[df_skills_job["label"]=="SOFT"]
print(len(df_hard_job), " Hard Skills collected")
print(len(df_soft_job), " Soft Skills collected")
```
