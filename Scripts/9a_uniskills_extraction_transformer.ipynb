# Prediction Uni Data


```python
import pandas as pd
import spacy
```

```python
df_full = pd.read_json('/content/drive/MyDrive/AML4TA/Project/Final_Uni_de.json')
df_full.head()
```

```python
trained_nlp = spacy.load("/content/drive/MyDrive/AML4TA/model-best")

collected_skills = []
doc = trained_nlp(df_full.loc[0,"translated_text"])
for ent in doc.ents:
    collected_skills.append({
        "skill": ent.text.lstrip('-').rstrip('-'),
        "label": ent.label_,
    })
collected_skills
```

```python
df_skills_uni = pd.DataFrame(collected_skills)
df_skills_uni
```

```python
df_skills_uni.to_json("/content/drive/MyDrive/AML4TA/Project/uni_datas_skills.json", orient="records")
```

```python
df_hard_uni = df_skills_uni[df_skills_uni["label"]=="HARD"] 
df_soft_uni = df_skills_uni[df_skills_uni["label"]=="SOFT"]
print(len(df_hard_uni), " Hard Skills collected")
print(len(df_soft_uni), " Soft Skills collected")
```
