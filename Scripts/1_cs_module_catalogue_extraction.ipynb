# Importing PDF and converting to text 

```python
import PyPDF2

def convert_pdf_to_text(file_path):
    with open(file_path, "rb") as file:
        # Create a PDF reader object
        pdf_reader = PyPDF2.PdfReader(file)

        # Extract text from all pages
        text = ""
        for page in range(len(pdf_reader.pages)):
            text += pdf_reader.pages[page].extract_text()

    return text

pdf_file_path = "csmodhbde.pdf"
text = convert_pdf_to_text(pdf_file_path)
print(text)
```

```python
# store as text file 

text_file_path = "csmodhbde.txt"
with open(text_file_path, "w") as file:
    file.write(text)
```

# Extracting skill sections from text file 

```python
def extract_text(file_path, start, end):
    with open(file_path, 'r') as file:
        content = file.read()

    start_index = 0
    results = []

    while True:
        start_index = content.find(start, start_index)
        if start_index == -1:
            break
        end_index = content.find(end, start_index)
        result = content[start_index:end_index]
        results.append(result)
        start_index = end_index

    return results

file_path = 'csmodhbde.txt'
start = 'Inhalte:'
end = 'Prüfungsleistung:'

texts = extract_text(file_path, start, end)
for text in texts:
    print(text)
```

# Observe most frequent words

```python
import spacy
from collections import Counter

# Load the German language model
nlp = spacy.load('de_core_news_lg')

# Load the JSON file
with open('csmodhbde.json', 'r') as f:
    data = json.load(f)

# Extract the text from the JSON data
if isinstance(data, dict):
    text = data['text']
elif isinstance(data, list):
    text = ' '.join(data)

# Process the text with spaCy
doc = nlp(text)

# Find the most common nouns
nouns = [token.lemma_.lower() for token in doc if token.pos_ == 'NOUN']
noun_freq = Counter(nouns)
most_common = noun_freq.most_common(30)

# Print the most common nouns and their frequencies
for word, frequency in most_common:
    print(f'{word}: {frequency}')

```
